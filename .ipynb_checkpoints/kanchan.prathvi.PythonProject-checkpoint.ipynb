```python
import requests                     # importing requests to hit the required url
import csv                          # import csv to create csv file
from bs4 import BeautifulSoup       # import beautifulSoup so as to pick up the tickers and other details from the source code
from itertools import zip_longest   # import zip_longest to store data in the csv file from the webpage

fileName = "stocks.csv"             # name of the csv file to be created
fileCsv = csv.writer(open(fileName, 'w' , newline='')) # opening the csv file in write mode
webPage = requests.get('https://money.cnn.com/data/hotstocks/') # the website url from which we have to extract data
webData = BeautifulSoup(webPage.text, 'html.parser') # scrapping the web page
extractWebData = webData.find_all(class_="wsod_dataTable wsod_dataTableBigAlt")  # extract the data present in class wsod_dataTable wsod_dataTableBigAlt

values = ['Most Actives', 'Gainers', 'Losers']  
value =[]  # creating an empty list
for webData in extractWebData:
    [subClass.extract() for subClass in webData.find_all(class_ = "wsod_aRight")]  # extracting class which contain two other subclasses
    abbrevation = [symbol.text for symbol in webData.find_all(class_= "wsod_symbol")]
    name = [span.text for span in webData.find_all("span")]
    value += [dict(zip(abbrevation, name))]  # store the value in dictionary using zip to store kep value pair   
displayData = dict(zip(values, value)) # contains ticker symbol and display data
tickerSymbol = [item for myList in zip_longest(*value) for item in myList if item] ## iterate using itertools.ziplongest
fileCsv.writerow([str(tickerSymbol)]) # write row in the csv file we created before

activeData = value[0]   #for most active
gainerData = value[1]   #for gainers
losersData = value[2]   #for losers
activeName = ''
gainerName = ''
losersName = ''

print('This is a program to scrape data from the https://money.cnn.com/data/hotstocks/ for a class project.\n' + '\nWhich stock are you interested in:\n')


print(values[0] + ':' + "\n")
for abbrevation, name  in activeData.items(): #print the names and ticker symbols of the Active Companies
    activeName += abbrevation +' '+ name + '\n'
print(activeName)


print('\n' + values[1] + ':' + '\n')    
for abbrevation, name  in gainerData.items():  #print the names and ticker symbols of the Gainer Companies
    gainerName += abbrevation +' '+ name + '\n'
print(gainerName)    


print('\n' + values[2] + ':' + '\n')    
for abbrevation, name  in losersData.items():      #print the names and ticker symbols of the Loser Companies
    losersName += abbrevation +' '+ name + '\n'
print(losersName)    

userInput = input("\nEnter your Interested Company :  ") #taking company symbol from user

if(userInput not in tickerSymbol): #checking whether the company symbol entered by user is valid or not
    print('Kindly Enter Valid Input') 
else:
    print('User inputs: ' + userInput + '\n') 
    webUrl = 'https://finance.yahoo.com/quote/' + userInput + '?p=' + userInput  
    page = requests.get(webUrl) 
    dataOne = BeautifulSoup(page.text, 'html.parser')  
    datasetOne = dataOne.find_all(class_="Bxz(bb) Bdbw(1px) Bdbs(s) Bdc($seperatorColor) H(36px)")  

    scrapedData=[]
    for data in datasetOne:  
        scrapedData += [span.text for span in data.find_all("span")] #flitering all the values in span

    PClose='' #strings for PClose
    Open=''   #strings for Open
    Volume='' #strings for Volume
    Cap=''    #strings for Cap
   
    for data in range(len(scrapedData)):
        if(scrapedData[data] == "Open"):    #scrap Open value of company
            Open = scrapedData[data+1] 
        if(scrapedData[data] == "Previous Close"): #scrap Previous Close of company
            PClose = scrapedData[data+1]
        if(scrapedData[data] == "Volume"):  #scrap volume of company 
            Volume = scrapedData[data+1]        
        if(scrapedData[data] == "Market Cap"): #scrap Market Cap value of company 
            Cap = scrapedData[data+1] 
            
    for item in value: 
        if userInput in item:   
            companyName= item[userInput] # retrieving the company name
            break;     

    #printing the required output
    print("\nThe Data for "+userInput+" "+ companyName+" is the following:" + "\n\n" + userInput +" "+companyName+ " \n" +"Previous Close" + ": "  + PClose + "\n" + "Open" + ": " + Open + "\n" + "Volume" + ": "  + Volume + "\n" + "Market Cap" + ": " + Cap)
```
